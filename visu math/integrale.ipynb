```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Paramètres
delta = 0.06  # taux d'actualisation
T = 5  # durée en années

# Revenu variable dans le temps
def f(t):
    return 500 + 20 * t

# Fonction à intégrer : revenu actualisé
def actualized_revenue(t):
    return f(t) * np.exp(-delta * t)

# Calcul de l'intégrale
valeur_actuelle, _ = quad(actualized_revenue, 0, T)

# Tracé de la fonction de revenu et du revenu actualisé
t_vals = np.linspace(0, T, 200)
revenu = f(t_vals)
revenu_actualise = revenu * np.exp(-delta * t_vals)

# Affichage graphique
plt.figure(figsize=(10, 6))
plt.plot(t_vals, revenu, label="Revenu instantané f(t)", linewidth=2)
plt.plot(t_vals, revenu_actualise, label="Revenu actualisé f(t)e^{-δt}", linestyle="--", linewidth=2)
plt.fill_between(t_vals, revenu_actualise, alpha=0.2, color='gray', label="Valeur actualisée totale")
plt.title(f"Revenu variable actualisé sur [0, {T}] ans\nValeur actuelle ≈ {valeur_actuelle:.2f} €")
plt.xlabel("Temps (années)")
plt.ylabel("€ par an")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
