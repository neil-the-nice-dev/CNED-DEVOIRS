```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Paramtres
delta = 0.06  # taux d'actualisation
T = 5  # dure en annes

# Revenu variable dans le temps
def f(t):
    return 500 + 20 * t

# Fonction  intgrer : revenu actualis
def actualized_revenue(t):
    return f(t) * np.exp(-delta * t)

# Calcul de l'intgrale
valeur_actuelle, _ = quad(actualized_revenue, 0, T)

# Trac de la fonction de revenu et du revenu actualis
t_vals = np.linspace(0, T, 200)
revenu = f(t_vals)
revenu_actualise = revenu * np.exp(-delta * t_vals)

# Affichage graphique
plt.figure(figsize=(10, 6))
plt.plot(t_vals, revenu, label="Revenu instantan f(t)", linewidth=2)
plt.plot(t_vals, revenu_actualise, label="Revenu actualis f(t)e^{-t}", linestyle="--", linewidth=2)
plt.fill_between(t_vals, revenu_actualise, alpha=0.2, color='gray', label="Valeur actualise totale")
plt.title(f"Revenu variable actualis sur [0, {T}] ans\nValeur actuelle  {valeur_actuelle:.2f} ")
plt.xlabel("Temps (annes)")
plt.ylabel(" par an")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
